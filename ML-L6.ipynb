```python
#Answer 1
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv('weather.csv')
```

```python
#Filling missing values (if any)
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
df.isnull().sum()
```

```python
#Describing the dataset
df.describe()
```

```python
#Now one by one making 20 different visualizations of the above dataset
```

```python
#Visualizatin 3----Rainfall
sns.distplot(df.Rainfall)

```

```python
#Visualization 4----Evaporation
sns.distplot(df.Evaporation)
```
